# Problem 1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0
    
    # Process each log entry
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        # Track peak usage
        if parked > peak:
            peak = parked
    
    print("Currently Parked Vehicles:", parked)
    
    if parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")
    
    print("Peak Parking Usage:", peak)
    

# Sample Test
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

# Problem 2: Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    eta = distance * 5
    
    # Traffic adjustment
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 10
    
    # Weather adjustment
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "stormy":
        eta += 20
    
    print("Estimated Delivery Time:", eta, "minutes")


# Sample Test
delivery_time_estimator(8, "High", "Rainy")
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy), "%")
    
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Test
theatre_occupancy(200, [1]*150)
```
